```python
#categories the data
from tensorflow.keras.utils import to_categorical
```

```python
y_example = to_categorical(y_train)
print(y_example,y_example.shape)
```

```python
y_example[0]
```

```python
y_cat_test = to_categorical(y_test,num_classes=10)
```

```python
y_cat_train = to_categorical(y_train,10)
```

```python
one_img.max(),one_img.min()
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
scaled_img = x_train[0]
scaled_img
```

```python
#reshape the scaled data
x_train = x_train.reshape(60000,28,28,1)
x_test = x_test.reshape(10000,28,28,1)
```

```python
x_train.shape,x_test.shape
```

#SPRINT2:MODEL BUILDING

```python
#Import libraries
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPool2D
```

```python
#Adding Layers
model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(4,4),activation='relu',input_shape=(28,28,1)))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
#Compilation of the model
model.compile(loss='categorical_crossentropy',optimizer='Adadelta',metrics=['accuracy'])
```

```python
#Early Stopping and Callbacks
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=1)
```

```python
#Test the Model
model.fit(x_train,y_cat_train,
          epochs=15,
          validation_data=(x_test,y_cat_test),
          callbacks=[early_stop])
```

Metrics:

```python
metrics = pd.DataFrame(model.history.history)
```

```python
metrics
```

```python
metrics.plot()
```

```python
metrics[['loss','val_loss']].plot()
```

```python
metrics[['accuracy','val_accuracy']].plot()
```

Evaluate the model


```python
model.evaluate(x_test,y_cat_test,verbose=0)
      #loss         |       #accuracy
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
predict_x=model.predict(x_test) 
classes_x=np.argmax(predict_x,axis=1)
```

```python
print(classification_report(y_test,classes_x))
```

```python
print(confusion_matrix(y_test,classes_x))
```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.heatmap(confusion_matrix(y_test,classes_x))
```
